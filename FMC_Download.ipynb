# ridersArrayLeaderboardData

```python
r = requests.get('https://www2.followmychallenge.com/live/tcrno9/data/ridersArrayLeaderboardData.json?1664536607')
r
```

```python
ridersArrayLeaderboardData = r.json()
len(ridersArrayLeaderboardData)
```

```python
rAdf=pd.DataFrame.from_dict(ridersArrayLeaderboardData)
rAdf
```

```python
rAdf[ rAdf["riderName"].str.contains('Strasser')][["riderName",'id']]
```

```python
params =rAdf[ rAdf["riderName"].str.contains('Strasser')][['imei']].to_dict(orient='records')[0]
params['id'] = params['imei']
```

```python
r = requests.get('https://www2.followmychallenge.com/live/tcrno9/application/get/get_route.php',params = params)
r
```

```python
coord=eval(r.text.split('"coordinates":')[1].split('}}}')[0])
coord[:10]
```

```python
import folium
```

```python
track = [c[::-1] for c in coord]
```

```python
m = folium.Map(tiles='OpenStreetMap')
folium.PolyLine(track).add_to(m)
m.fit_bounds(m.get_bounds())
#[CircleMarker(l,color='red',popup=popup(Bibs[1],t,sticky=True),radius=1).add_to(m) for i,l,t in route[names[1]]]
m
```
